### I only want to use the 50 US states and Washington D.C. for my data and houses that have a status of sold.
I will begin cleaning the data to only contain the states by removing the US territories, after that I will remove any houses that are not currently sold so that we can begin looking at the data.


```python
not_wanted = ["Puerto Rico", "Virgin Islands", "Guam", "New Brunswick"] #selecting locations to remove


fifty_states = realtor_data[~realtor_data['state'].isin(not_wanted)] #removing the locations not wanted

fifty_states_sold = fifty_states[fifty_states['status'] == 'sold'] #removing the houses that are not currently sold

fifty_states_sold.head()
```
